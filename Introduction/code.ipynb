```python
## Import de la librairie Random
import random
import numpy as np
import math

##Initialisation des variables
MAXGEN = 100 # Nombre maximum de générations
POPSIZE = 100 # La taille de la population
N = 20 # la longueur d'un génome
PMUT = 0.61 # probabilité d'une mutation
PX = 0.6 # probabilité d'une recombination
pop = []
f = []
Cf = []
nb = []
```

```python
##Fct d'initialisation (création de la première generation de manière aleatoire)
def initialisation():
  nb = [0]*N
  pop = ['']*POPSIZE
  for i in range(POPSIZE):
    for j in range(N):
      if j <= 20:
        nb[j] = random.randint(0, 1)
        pop[i] = pop[i] + str(nb[j])
  return pop
```

```python
##Fct de fitness, le but est de compter le nombre de "1" par individu
def fitness(pop):
  f = [0]*POPSIZE
  for i in range(POPSIZE):
    f[i] = pop[i].count("1")
    #print(pop[i]," ",f[i])
  return pop, f
```

```python
## Fct qui calcule le Cf
def generateCf(pop):
  Cf = [0]*POPSIZE
  res = fitness(pop)
  pop = res[0]
  f = res[1]

  for j in range(POPSIZE):
    if j == 0 :
      Cf[j] = f[j]
    else :
      Cf[j] = Cf[j-1] + f[j]  
  return Cf

##Fct qui procède a la selection des individus
def selection(pop):

  Cf = generateCf(pop)
  liste = []
  for i in range(100):
      liste.append(random.randint(0, Cf[POPSIZE-1]))

  print(liste, " ", len(liste))

  selectarray = []
  listCf = []

  for y in range(len(liste)):
    for t in range(POPSIZE):
      if liste[y] <= Cf[t] :
          selectarray.append(pop[t])
          listCf.append(Cf[t])
          break

  ##Print de test
  for o in range(5):
    print("individu : ", selectarray[o], "   Nombre aleatoire : " , liste[o] , "   Cfselectionné : " , listCf[o])
  return selectarray
```

```python
def mutation(ind1):
    ind = list(ind1)
    for i in range(len(ind)):
        if random.random() < PMUT:
            #nbmut = nbmut+1
            if ind[i] == '1':
                ind[i] = '0'
            else:
                ind[i] = '1'
    ind1 = ''.join(ind)
```

```python
def crossover(ind1,ind2):
  ##On génére un intier aléatoire entre 0 et 19
  positioncross = random.randint(0, 19)
  #nbcombi+=1
  ##Echange de genome a droite en fonction de la position du crossingover
  gen1  = ind1[0:positioncross]+ind2[positioncross:]
  gen2 = ind2[0:positioncross]+ind1[positioncross:]
  print("parent1 ",ind1,"parent2 : ",ind2,"enfant1 : ",gen1,"enfant2 : ",gen2 )
```

```python
print("Max gen : ",MAXGEN)
print("Taille de la population :", POPSIZE)
print("longueur du genome :", N)
print("Probabilité de mutation :", PMUT)
print("Probabilité d'une recombination :",PX)
```

```python
random.seed()
nb = [0]*N
pop = ['']*POPSIZE
##Initialisation
for i in range(POPSIZE):
  for j in range(N):
    if j <= 20:
      nb[j] = random.randint(0, 1)
      pop[i] = pop[i] + str(nb[j])
      
for g in range(MAXGEN):
  pop = selection(pop)
  for ind in pop:
   mutation(ind)
  for i in range(0,POPSIZE,2):
    if random.random() < PX:
      crossover(pop[i],pop[i + 1])
```
